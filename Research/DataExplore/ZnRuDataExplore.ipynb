```python
import matplotlib.pyplot as plt

def set_shared_ylabel(a, ylabel, labelpad = 0.01):
    """Set a y label shared by multiple axes
    Parameters
    ----------
    a: list of axes
    ylabel: string
    labelpad: float
        Sets the padding between ticklabels and axis label"""

    f = a[0].get_figure()
    f.canvas.draw() #sets f.canvas.renderer needed below

    # get the center position for all plots
    top = a[0].get_position().y1
    bottom = a[-1].get_position().y0

    # get the coordinates of the left side of the tick labels 
    x0 = 1
    for at in a:
        at.set_ylabel('') # just to make sure we don't and up with multiple labels
        bboxes, _ = at.yaxis.get_ticklabel_extents(f.canvas.renderer)
        bboxes = bboxes.inverse_transformed(f.transFigure)
        xt = bboxes.x0
        if xt < x0:
            x0 = xt
    tick_label_left = x0

    # set position of label
    a[-1].set_ylabel(ylabel)
    a[-1].yaxis.set_label_coords(tick_label_left - labelpad,(bottom + top)/2, transform=f.transFigure)
 
# data to plot
n_groups = len(deli_date)
 
# create plot
f,a = plt.subplots(2, sharex=True, gridspec_kw={'hspace':0})
index = np.arange(n_groups)
bar_width = 0.35
opacity = 1.0
 
# ru_fig
ru_rects1 = a[0].bar(index, df_ru['ru_price_num'], 
                    bar_width, alpha=opacity, color='b',label='ru_price_num')
 
ru_rects2 = a[0].bar(index + bar_width, df_ru['ru_price_chg_num'], 
                    bar_width,alpha=opacity,color='g',label='ru_price_chg_num')
a[0].legend()

# zn_fig
zn_rects1 = a[1].bar(index, df_zn['zn_price_num'], 
                   bar_width, alpha=opacity, color='b', label='zn_price_num')
 
zn_rects2 = a[1].bar(index + bar_width, df_zn['zn_price_chg_num'], 
                    bar_width,alpha=opacity,color='g',label='zn_price_chg_num')
a[1].legend()

plt.xticks(index + bar_width, df_ru.index.values)

set_shared_ylabel(a, 'Number of data ')

f.set_size_inches(18.5, 12.5)
f.savefig('Ru_Zn_data.png', dpi=100)
plt.show()
```
